```python
x = np.linspace(0, 10, 1000)

ax = plt.gca()
ax.plot(x, np.sin(x),   color='blue',    linestyle='solid')   # full color name (salmon - лососевый)   # -
ax.plot(x, np.sin(x-1), color='g',       linestyle='dashed')  # single letter                          # --
ax.plot(x, np.sin(x-2), color='0.2'   ,  linestyle='dashdot') # grayscale (0=black, 1=white)          # -.
ax.plot(x, np.sin(x-3), color='#FFDDAA', linestyle='dotted')  # hex RGB triplet                        # :
ax.plot(x, np.sin(x-4), color=(1.0, 0.3, 0.3))                # RGB tuple (0–1)
ax.plot(x, np.sin(x-5), '--y')                                # mixed approach
 
plt.show()
```

```python
fig, ax = plt.subplots(5)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))
ax[2].plot(x, np.sin(x))
ax[3].plot(x, np.sin(x))
ax[4].plot(x, np.sin(x))

ax[1].set_xlim(-2, 12)
ax[1].set_ylim(-2, 2)

ax[2].set_xlim(12, -2) # задаёт границы по X. Если левый > правый, ось разворачивается.
ax[2].set_ylim(2, -2)  # mirroring the graph

#ax[3].axis('tight') # автоматический подбор 'лучшего' варианта
ax[3].autoscale(tight=True) # убирает все поля вокруг данных (границы осей = min и max данных)

ax[4].axis('equal') # делает масштаб осей одинаковым, что искажает внешний вид, но сохраняет геометрические пропорции.

plt.tight_layout() # графики не наезжают друг на друга (автоматически подбирается)

plt.show()
```

```python
plt.subplot(3, 1, 1)
plt.plot(x, np.sin(x) * np.sqrt(2))
plt.title('Синус')
plt.xlabel('x')
plt.ylabel('$\sqrt{2}$ sin(x)') # support Latex

plt.subplot(3, 1, 2)
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.legend()
plt.title('Синус и косинус')

plt.subplot(3, 1, 3)
plt.plot(x, np.sin(x), '--k', label='$\sin x$')
plt.plot(x, np.cos(x), '-r',  label='$\cos x$' )
plt.legend()
plt.title('Синус и косинус 2')

plt.subplots_adjust(hspace=1) # or auto-version plt.tight_layout()

plt.show()
```

```python
rng = np.random.default_rng(1)
x = np.linspace(0, 10, 100)
dy = 0.5
y = np.sin(x) + dy * rng.random(100)

plt.plot(x,y, '--k')

plt.errorbar(x, y, yerr=dy, fmt='.g')

plt.fill_between(x, y - dy, y + dy, color='0.75', alpha=0.7)

plt.show()
```

```python
# графики плотности и контурные графики (на 2D изобразить 3D)
x = np.linspace(0, 6, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y) # превращает одномерные массивы координат (x, y) в двумерные матрицы, которые вместе задают все точки прямоугольной сетки.
print(X, end='\n\n') # каждая строка повторяет массив x (40 строка на 50 столбцов)
print(Y) # каждый столбец повторяет массив y (40 на 50).

# Таким образом, пара (X[i, j], Y[i, j]) даёт координаты точки на пересечении j-го значения x и i-го значения y.
```

```python
def f(x, y):
    return np.sin(x) ** 4 + np.cos(y + x * 30) * np.sin(y)

Z = f(X, Y)

print(Z.shape)
```

```python
# plt.contour(X, Y, Z)
# plt.contour(X, Y, Z, colors='black')
# plt.contourf(X, Y, Z, 10, cmap='RdGy_r')
# plt.contourf(X, Y, Z, 10, cmap='RdGy_r')
# plt.contourf(X, Y, Z, 10, cmap='RdGy')
cnt = plt.contour(X, Y, Z, 10, colors='black', alpha=0.5)
plt.clabel(cnt, inline=True)

plt.colorbar()

plt.imshow(
    Z,
    extent=[0,6, 0,5],
    origin='lower',
    cmap='RdGy',
    interpolation='gaussian',
    aspect='equal',
    alpha=1
)

plt.show()
```

```python
# Гистограммы
rng = np.random.default_rng(1)
data = rng.normal(size=1000)

plt.hist(data, bins=30, density=True)

plt.show()
```

```python
rng = np.random.default_rng(1)
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)

p = dict(alpha=0.3, bins=40)

plt.hist(x1, **p)
plt.hist(x2, **p)
plt.hist(x3, **p)

plt.show()
```
