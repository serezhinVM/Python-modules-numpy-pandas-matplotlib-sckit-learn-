```python
x = np.linspace(0, 10, 1000)

ax = plt.gca()
ax.plot(x, np.sin(x),   color='blue',    linestyle='solid')   # full color name (salmon - лососевый)   # -
ax.plot(x, np.sin(x-1), color='g',       linestyle='dashed')  # single letter                          # --
ax.plot(x, np.sin(x-2), color='0.2'   ,  linestyle='dashdot') # grayscale (0=black, 1=white)          # -.
ax.plot(x, np.sin(x-3), color='#FFDDAA', linestyle='dotted')  # hex RGB triplet                        # :
ax.plot(x, np.sin(x-4), color=(1.0, 0.3, 0.3))                # RGB tuple (0–1)
ax.plot(x, np.sin(x-5), '--y')                                # mixed approach

ax.axis('equal')

plt.show()
```

```python
# графики плотности и контурные графики (на 2D изобразить 3D)
x = np.linspace(0, 6, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y) # превращает одномерные массивы координат (x, y) в двумерные матрицы, которые вместе задают все точки прямоугольной сетки.
print(X, end='\n\n') # каждая строка повторяет массив x (40 строка на 50 столбцов)
print(Y) # каждый столбец повторяет массив y (40 на 50).

# Таким образом, пара (X[i, j], Y[i, j]) даёт координаты точки на пересечении j-го значения x и i-го значения y.
```

```python
def f(x, y):
    return np.sin(x) ** 4 + np.cos(y + x * 30) * np.sin(y)

Z = f(X, Y)

print(Z.shape)
```

```python
# plt.contour(X, Y, Z)
# plt.contour(X, Y, Z, colors='black')
# plt.contourf(X, Y, Z, 10, cmap='RdGy_r')
# plt.contourf(X, Y, Z, 10, cmap='RdGy_r')
# plt.contourf(X, Y, Z, 10, cmap='RdGy')
cnt = plt.contour(X, Y, Z, 10, colors='black', alpha=0.5)
plt.clabel(cnt, inline=True)

plt.colorbar()

plt.imshow(
    Z,
    extent=[0,6, 0,5],
    origin='lower',
    cmap='RdGy',
    interpolation='gaussian',
    aspect='equal',
    alpha=1
)

plt.show()
```

```python
from sklearn.datasets import load_digits

digits = load_digits(n_class=6)

fig, ax = plt.subplots(8, 8)
for i, ax_ in enumerate(ax.flat):
    ax_.imshow(digits.images[i], cmap='binary')
    ax_.set(xticks=[], yticks=[])
```

```python
from sklearn.manifold import Isomap

iso = Isomap(n_components=2, n_neighbors=10)
prj = iso.fit_transform(digits.data)

plt.scatter(
    prj[:, 0],
    prj[:, 1],
    c=digits.target,
    cmap=plt.get_cmap('jet', 6)
)

plt.colorbar(ticks=range(6))
plt.clim(-0.5, 5.5)

plt.show()
```
