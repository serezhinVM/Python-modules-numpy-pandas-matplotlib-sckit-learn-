```python
# Pandas - удобная надстройка над NumPy. В NumPy индексы - целые числа, в Pandas - всё что угодно (время, строки и т.д.).
# Series
# DataFrame
# Index
import numpy as np
import pandas as pd
```

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0]) #  Если упростить до предела: Series = ndarray + Index
print(data)

print(data.values)
print(type(data.values))

print(data.index)
print(type(data.index))

print(data[1])
print(data[1:3])
```

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0], index = ['a', 'b', 'c', 'd'])
print(data)

print(type(data.index[0]))
print(data['a'])
print(data['b':'d']) # the last element is included (not like with inreger slices)
```

```python
# Index - способ сослаться (reference) на данные в Series или DataFrame
index = pd.Index([2, 5, 3, 5, 71])
print(index)
print(type(index))

print(index[1])
print(index[::2])

# index[1] = 21 - Index is an immutable object!
```

```python
# Index is familiar with a built-in Python's set
index1 = pd.Index([2, 5, 3, 5, 71])
index2 = pd.Index([1, 2, 5, 71, 4])

print(index1.intersection(index2)) # &
print(index1.union(index2)) # |
print(index1.symmetric_difference(index2)) # ^ and etc.
```

```python
# Deleting Nan values from Series
x = pd.Series([1, np.nan, None, 5, pd.NA, 'hello'])
print(x)

print(x.isnull())
print(x.notnull(), end='\n\n')
print(x[x.notnull()], end='\n\n') # 1. masking

print(x.dropna(), end='\n\n') # 2. drop out Nan values
```

```python
# Deleting Nan values from DataFrame
x = pd.DataFrame([[pd.NA, np.nan, None], [1,2,3], [2,np.nan,3]])
print(x, end='\n\n')
print(x.dropna(axis=0), end='\n\n') # Non-zero rows
print(x.dropna(axis=1), end='\n\n') # Non-zero columns

print(x.dropna(axis=0,how='all')) # default settings how = 'any'
```

```python
x = pd.DataFrame([[pd.NA, np.nan, None], [1,2,3], [2,np.nan,3]])
print(x.dropna(axis=1, thresh=2)) # должно быть минимум N непустых значений, thresh - молотить
```

```python
# how = all - убрать, если все значения отсутствуют
# how = any - убрать, если хотя бы одно значение отсутствуют
# tresh = n - убрать, если хотя бы n значений отсутствуют
```

```python
x = pd.DataFrame([None, 1000, np.nan, None, 1, 2, 3], dtype='Int32')
print(x, end='\n\n')

print(x.fillna(-999), end='\n\n') # fillna(n) - заменить nan на n
print(x.ffill(), end='\n\n') # Заполнить следующее предыдущим значением, forward
print(x.bfill(), end='\n\n') # Заполнить предыдущее следующим, back
```

```python
# Иерархическая индексация - Hierarchical indexing or Multindexing
index = [
    ('A1', 2025),
    ('A1', 2026),
    ('A2', 2025),
    ('A2', 2026),
    ('A3', 2025),
    ('A3', 2026),
]

data = [1, 2, 3, 4, 5, 6]
s = pd.Series(data, index=index)
print(s, end = '\n\n')
print(s[[0,2,4]], end = '\n\n')

print(s[[i for i in s.index if i[0] == 'A1']], end = '\n\n') 
```

```python
# В index содержится иерархия, которую можно реализовать через Multiindex
mi = pd.MultiIndex.from_tuples(index)
print(mi)

s = s.reindex(mi) # способ перестроить существующий объект (Series или DataFrame) так, чтобы он соответствовал новому индексу mi.
print(s)
```

```python
s1 = pd.Series(data, index=mi) # способ создать новый перестроенный объект
print(s1)
```

```python
# Ways how to create MultiIndex
# 1. from_tuples
index =[
    ('A1', 2025),
    ('A1', 2026),
    ('A2', 2025),
    ('A2', 2026)
]
    
index = pd.MultiIndex.from_tuples(index)
print(index)
```

```python
# 2. from_arrays
index = pd.MultiIndex.from_arrays([['A1', 'A1', 'A2', 'A2'], [2025, 2026, 2025, 2026]])
print(index)
```

```python
# 3. from_product
index = pd.MultiIndex.from_product([['A1', 'A2'], [2025, 2026]])
print(index)
```

```python
# codes
index = pd.MultiIndex(
    levels = [['A1', 'A2'], [2025, 2026]],
    codes = [[0,0,1,1], [0,1,0,1]]
)
print(index)
```
