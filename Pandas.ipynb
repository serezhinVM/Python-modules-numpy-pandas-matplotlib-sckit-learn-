```python
# Pandas - удобная надстройка над NumPy. В NumPy индексы - целые числа, в Pandas - всё что угодно (время, строки и т.д.).
# Series
# DataFrame
# Index
import numpy as np
import pandas as pd
```

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0]) #  Если упростить до предела: Series = ndarray + Index
print(data)

print(data.values)
print(type(data.values))

print(data.index)
print(type(data.index))

print(data[1])
print(data[1:3])
```

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0], index = ['a', 'b', 'c', 'd'])
print(data)

print(type(data.index[0]))
print(data['a'])
print(data['b':'d']) # the last element is included (not like with inreger slices)
```

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0], index = [1, 10, 7, 'd'])
print(data)

print(type(data.index))
print(type(data.index[0]))
print(type(data.index[3]))
print(data.index)

print(data[1])
print(data['d'])
print(data[10:'d'])
```

```python
my_dict = {
    'A': 10,
    'B': 20,
    'C': 30,
    'D': 40,
    'E': 50
}

data_dict = pd.Series(my_dict)
print(data_dict)

print(data_dict['b':'d']) # 'dict' with slices
```

```python
# Conclusions about how to create Series?

# 1. Arrays
data = pd.Series([0.25, 0.5, 0.75, 1.0])
print(data)

data = pd.Series(5, index=[10, 20, 30])
print(data, end='\n\n')

# 2. Dictionaries
data_dict = pd.Series(my_dict)
print(data_dict)
```

```python
# DataFrame - n-dimensional massive, Series - 1-d massive. DataFrame consists of some Series
my_dict1 = {
    'A': 10,
    'B': 20,
    'C': 30,
    'D': 40,
    'E': 50,
   #'N' : 60
}

my_dict2 = {
    'A': 11,
    'B': 21,
    'C': 31,
    'D': 41,
    'E': 51,
    'V': 60
}

data_dict1 = pd.Series(my_dict1)
print(data_dict1)

data_dict2 = pd.Series(my_dict2)
print(data_dict2)


df = pd.DataFrame({'dict_01': data_dict1, 'dict_02': data_dict2})
print(df)

print(df.values)
print(type(df.values))

print(df.columns)
print(type(df.columns))

print(df.index)
print(type(df.index))

print(df['dict_01'])
print(df['dict_02'])
```

```python
df = pd.DataFrame({'dict_01': data_dict1}) # Series - 1-d massive, but DataFrame is 2-d (analogy with numpy: array and .reshape)
print(df)
print(df.values)

df = pd.DataFrame(data_dict1, columns=['rrr'])
print(df)
```

```python
print(pd.DataFrame([{'a': i, 'b': 2 * i} for i in range(4)]))

# NaN - Not a Number
print(pd.DataFrame(np.zeros(3)))
```

```python
# Index - способ сослаться (reference) на данные в Series или DataFrame
index = pd.Index([2, 5, 3, 5, 71])
print(index)
print(type(index))

print(index[1])
print(index[::2])

# index[1] = 21 - Index is an immutable object!
```

```python
# Index is familiar with a built-in Python's set
index1 = pd.Index([2, 5, 3, 5, 71])
index2 = pd.Index([1, 2, 5, 71, 4])

print(index1.intersection(index2)) # &
print(index1.union(index2)) # |
print(index1.symmetric_difference(index2)) # ^ and etc.
```

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0], index = ['a', 'b', 'c', 'd'])
print(data['a'])

print(data)
print('a' in data)
print('a1' in data)

data['a'] = 99
data['a1'] = 999
print(data)

print(data['c': 'a1']) # explicit slice
print(data[2:]) # Absolutely the same (implicit slice)
print(data[(data > 0.3) & (data < 0.8)])

print(data[['c', 'a']]) # print(data[[2, 0]])
#data.iloc - keys, data.loc - values
```

```python
rng = np.random.default_rng(seed=1)
s = pd.Series(rng.integers(0, 10, 6))
print(s, end='\n\n')

print(np.exp(s), end='\n\n')

df = pd.DataFrame(rng.integers(0, 10, (3,4)), columns=['A','B','C','D'])
print(df, end='\n\n')
print(np.sin(df * 4), end='\n\n')
```

```python
# Индексы не просто сохраняются, а согласовываются
my_dict1 = {
    'A': 10,
    'B': 20,
    'C': 30,
    'D': 40,
    'E': 50,
   #'N' : 60
}

my_dict2 = {
    'A': 11,
    'B': 21,
    'C': 31,
    'D': 41,
    'E': 51,
    'V': 60
}

df = pd.DataFrame({'dict_01': data_dict1, 'dict_02': data_dict2})
print(df, end='\n\n')

print(data_dict1 + data_dict2, end='\n\n')
print(np.add(data_dict1, data_dict2), end='\n\n')
print(data_dict1.add(data_dict2, fill_value=5), end='\n\n') # fill_value can replace Nans
```

```python
print(df, end='\n\n')
df1 = pd.DataFrame({'dict_02': data_dict1, 'dict_03': data_dict2})
print(df1, end='\n\n')
print(df + df1, end='\n\n') #Lot's of Nan
print(df.add(df1, fill_value = 5), end='\n\n')
```

```python
A = rng.integers(10, size=(3,4))
print(A, end='\n\n')
print(A[0], end='\n\n')
print(A - A[0], end='\n\n')

df = pd.DataFrame(A, columns=['A','B','C','D'])
print(df, end='\n\n')
print(df - df.iloc[0], end='\n\n')
print(df.subtract(df['A'], axis=0), end='\n\n')
```
